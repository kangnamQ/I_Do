## 데이터 파일 읽기 예

```python
import numpy as np
import pandas as pd

unames = ['UserID', 'Gender', 'Age', 'Occupation', 'Zip']
rnames = ['UserID', 'MovieID', 'Rating', 'Timestamp']
mnames = ['MovieID', 'Title', 'Genres']

users = pd.read_csv('movielens/users.dat', sep = '::', header = None, names = unames, engine = 'python')
ratings = pd.read_csv('movielens/ratings.dat', sep = '::', header = None, names = rnames, engine = 'python')
movies = pd.read_csv('movielens/movies.dat', sep = '::', header = None, names = mnames, engine = 'python')
```

```python
users[:5]
```

```python
ratings[:5]
```

```python
movies[:5]
```

```python
ratings.tail(3)
```

```python
#merge : 데이터 프레임을 병합하는 명령어
data = pd.merge(pd.merge(ratings, users), movies)
```

```python
data[:5]
```

```python
# loc은 행을 나타낸다.
data.loc[0]
```

```python
#성별 평점의 피벗 테이블을 보여준다.
mean_ratings = data.pivot_table('Rating', index='Title', columns='Gender', aggfunc='mean')

# data.pivot_table?
# data.pivot_table(values=None, index=None, columns=None, aggfunc='mean',
#                  fill_value=None, margins=False, dropna=True, margins_name='All')
# index : index 색인으로 사용될 컬럼
# columns : column 색인으로 사용될 컬럼
# aggfunc : 데이터 축약 시 사용할 함수 (mean, sum, count ...)
# margins : 행/열별 총 합

mean_ratings[:5]
```

```python
#영화제목으로 그룹화
#size()함수를 사용하여 제목별 평점의 건 수 조회 (시리즈 객체를 리턴)
ratings_by_title = data.groupby('Title').size()

ratings_by_title[:10]
```

```python
# 평점의 건수가 250개 이상되는 영화제목만 나오게 / == 수식적용 가능
active_titles = ratings_by_title.index[ratings_by_title >= 250]

active_titles
```

```python
#평점 건수가 250이상인 성별 평점의 피벗 테이블
mean_ratings = mean_ratings.loc[active_titles]

mean_ratings[:5]
```

```python
#여성에게 인기 있는 TOP 5 영화
top5_female_ratings = mean_ratings.sort_values(by = 'F', ascending = False)

top5_female_ratings[:5]
```

```python
#여성이 남성보다 더 선호하는 TOP 5 영화
#남성에서 여성의 평점을 뺴여 평점차이를 diff로 나타냄.

mean_ratings['Diff'] = mean_ratings['M'] - mean_ratings['F']

#평점 차이를 올림 차순으로 정렬함.
sorted_by_diff = mean_ratings.sort_values(by = 'Diff')

sorted_by_diff[:5]
```

```python
#올림 차순의 역순, 즉 내림 차순 정렬임
#즉 남성이 여성보다 더 선호하는 Top5임
sorted_by_diff[::-1][:5]
```

```python
#평점의 표준편차가 큰 영화를 250개이상만 선택하여 TOP5선택

rating_std_by_title = data.groupby('Title')['Rating'].std()

rating_std_by_title = rating_std_by_title.loc[active_titles]

rating_std_by_title.sort_values(ascending = False)[:5]
```
