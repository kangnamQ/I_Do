```python
import urllib.request
import re
from bs4 import BeautifulSoup
import pandas as pd
import sqlite3
```

```python
page1 = urllib.request.urlopen("http://ncov.mohw.go.kr/").read()
soup1 = BeautifulSoup(page1, 'html.parser')

patient_data = soup1.find('div', {'class' : 'container'})
current_data = patient_data.find('div', {'class' : 'liveNumOuter'})
```

```python
when_current = current_data.find('span', {'class' : 'livedate'}).text  
print(when_current)

#split으로 나누기
when_current_t = re.split('\W+', when_current)
print(when_current_t)

# 합쳐서 문장 만들기.
when = (when_current_t[1]+'월 '+when_current_t[2]+'일 '+when_current_t[3]+' '+when_current_t[4])
print(when)
```

```python
#확진자 수 가져오기
today_current = current_data.find('div', {'class' : 'liveNum_today_new'})

tit = today_current.find('strong', {'class' : 'tit'}).text
subtit = today_current.find('span', {'class' : 'subtit'}).text

subtit_number = today_current.find('span', {'class' : 'data'}).text

# 합쳐서 문장 만들기.
when_current_info= (when+' '+tit+' 중 '+subtit+'자는 '+subtit_number+"명 입니다. \n")
print(when_current_info)
```

```python
page2 = urllib.request.urlopen("http://ncov.mohw.go.kr/bdBoardList_Real.do?brdId=1&brdGubun=13&ncvContSeq=&contSeq=&board_id=&gubun=").read()
soup2 = BeautifulSoup(page2, 'html.parser')
```

```python
#table_today = soup2.find('table',{'class':'num midsize'})
data = []

tables = soup2.select('table')
table_html = str(tables)
table_df_list = pd.read_html(table_html)
table_df = table_df_list[0]

table_df
```

```python
#먼저 0번항목에 합계가 있으니 시도별 금일 확진자의 정보를 봅니다.
table_df_0_sum = table_df.loc[0]
table_df_0_sum

#아마도 한글이라서 깨지는 걸까.. html을 가져와서 꺠지는걸까
#깨지는 이유를 확실하게 모르겠습니다.. 
```

```python
#합계와 검열을 제외한 나머지 도시를 보겠다.
table_df_p = table_df.drop([0,18])
table_df_p
```

```python
# DataFrame 멀티색인
# 기본 색인 
# xs() 메서드를 활용한 색인 : axis=0, level=0 (default)
#   - axis=0 (row),  axis=1 (column)
#   - level=0 (상위 레벨),  level = 1 (하위 레벨)
#   - xs() 메서드로 column색인 시 축 설정 필요(axis가 default로 0이므로)
#   - 특정 레벨 or 특정 축만 가능하고 동시 선택은 불가
#출처: https://data-make.tistory.com/126 [Data Makes Our Future]


# Multiindex라고 계층적 인덱스로 긇혀서 일단은 3부분으로 나누어보았다.
# stack이라는 함수도 있었지만 뭔가 잘되지않아 일단은 일일히 해주었다.
# table_df_p1.stack(level=-1,dropna=True)

#print(table_df_p1) -> print로하면 글자처럼 나오고 그냥 쓰면 표같이 나온다.
#근데 print 말고는 3개를 다 띄울 방법이 마땅치않아 한칸씩 배정해주기로 하였다.

table_df_p1 = table_df_p.xs('시도명', axis=1)
table_df_p1
```

```python
table_df_p2 = table_df_p.xs('전일대비확진환자 증감', axis=1)
table_df_p2
```

```python
table_df_p3 = table_df_p.xs('확진환자 (명)', axis=1)
table_df_p3
```

```python
#table_df_pa = pd.merge(pd.merge(table_df_p1, table_df_p1), table_df_p3)
# => merge는 오류가 뜬다...

#merge()함수는 두 데이터프레임을 각 데이터에 존재하는 고유값(key)을 기준으로 병합할때 사용한다.
#pd.merge(df_left, df_right, how='inner', on=None)이 default이다.
#따라서 이경우 같은 key값이 없어 merge가 되지 않았다.!

# 따라서 concat이란 함수를 써서 합쳐주어서 표 형식으로 만들었다.
table_df_pa = pd.concat([table_df_p1,table_df_p2, table_df_p3],axis=1)
table_df_pa
```

```python
#확진환자가 높은순으로 볼 수 있을까?

table_df_pa.sort_values(by=['확진환자'], ascending=False)
```

```python
#근데 너무 많이 나오니 이런식으로 Top5를 볼 수 있다.
Most_Patient = table_df_pa.sort_values(by=['확진환자'], ascending=False)
Most_Patient[:5]
```

```python
#그렇다면 확진환자가 가장 없는 Top3은 어디일까?

Most_L_P = table_df_pa.sort_values(by=['확진환자'], ascending=True)
Most_L_P[:3]
```

```python
#금일 확진자가 가장 많이 나온 곳은 어디일까?
#국내발생도 결과는 같으니 생략

M_H_P_T = table_df_pa.sort_values(by=['합계'])
M_H_P_T[::-1][:1]
```

```python
#그렇다면 금일 확진자가 가장 적게 나온곳은 어디일까?

Most_L_P = table_df_pa.sort_values(by=['합계'])
Most_L_P[:1]
```

```python
#그렇다면 발생률이 가장 높은 Top3는 어디일까?

Most_Per = table_df_pa.sort_values(by=['발생률 (*)'], ascending=False)
Most_Per[:3]
```

```python
#그럼 이제 한번 그래프로 표시해보자

table_df_pa1 = pd.concat([table_df_p1,table_df_p2],axis=1)
table_df_pa1
```

```python
#그럼 이제 한번 그래프로 표시해보자

#확진환자와 발생률 (*)를 제거하고 시작하겠다.

table_df_pa2 = pd.concat([table_df_p1,table_df_p3],axis=1)
table_df_pa2 = table_df_pa2.drop('확진환자', axis=1, inplace=False)
table_df_pa2 = table_df_pa2.drop('발생률 (*)', axis=1, inplace=False)
table_df_pa2
```

```python
#이전 Final 프로젝트에서 계속해서 오류나던것을 해결함...

#폰트깨짐으로 폰트를 다시 설정해줌
import matplotlib.font_manager as fm
font_list = fm.findSystemFonts(fontpaths = None, fontext = 'ttf')
fontnow = fm.FontProperties(fname=font_list[0])
from matplotlib import pyplot as plt
import matplotlib

font_location = "C:\jupyter\Pandas\malgun.ttf"
font_name = fm.FontProperties(fname = font_location).get_name()
plt.rc('font', family='Malgun Gothic')


ax = table_df_pa1.plot(kind='barh', stacked=True, title="COVID19 시도별 발생동향", rot=0)
for p in ax.patches:
    left, bottom, width, height = p.get_bbox().bounds
    ax.annotate(width, xy=(left+width/2, bottom+height/2), ha='center', va='center')
plt.box(False)
# => 데이터가 합쳐지지 않음
# => 따라서 데이터를 다시 나누어서 그려야함.

plt.xlabel("전일 대비 확진환자")
plt.ylabel("도시")
plt.title("금일 COVID19 확진 현황")


plt.show()
```
