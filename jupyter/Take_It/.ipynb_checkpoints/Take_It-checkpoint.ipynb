## Take It

```python
import pygame
from random import randrange

#색
black = (0,0,0)
white = (255,255,255)

point_list = ["plus.png", "minus.png"]
item_list = ["heal.png", "fast.png", "slow.png", "stop.png"]

#떨어지는 class
class Plus(pygame.sprite.Sprite):
    def __init__(self, x, y):
        pygame.sprite.Sprite.__init__(self)
        self.image = pygame.image.load("plus.png").convert()
        self.image.set_colorkey(white)
        self.rect = self.image.get_rect()
        
        self.rect.x = randrange(0,screen_width)
        self.rect.y = randrange(-130,-30)
        
        def reset_pos(self):
            self.rect.x = randrange(0,screen_width)
            self.rect.y = randrange(-130,-30)
            
        def update(self, change_y):
            self.rect.y += change_y
            if self.rect.y > screen_height:
                self.reset_pos()
                               
class Player(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        
        self.width = 50
        self.height = 50
        
        self.image = pygame.image.load("hand.png").convert()
        self.image.set_colorkey(white)
        
        self.rect = self.image.get_rect()
        
        self.screenheight = pygame.display.get_surface().get_height()
        self.screenwidth = pygame.display.get_surface().get_width()
        self.rect.topleft = (0,self.screenheight-self.height)
        
        def update(self):
            pos = pygame.mouse.get_pos()
            self.rect.left = pos[0]
        
            if self.rect.left > self.screenwidth - self.width:
                self.rect.left = self.screenwidth - self.width
                
                
#pygame 초기화                
pygame.init()

#창 너비 지정 및 창크기 설정
screen_width = 800
screen_height = 600
screen = pygame.display.set_mode([screen_width,screen_height])
pygame.display.set_caption('Take It')


#이미지 및 사운드 load
background = pygame.Surface(screen.get_size())
background.fill(black)
background_image = pygame.image.load("library.png").convert()
background_position=[0,0]

player_image = pygame.image.load("hand.png").convert()
player_image.set_colorkey(white)

back_sound = pygame.mixer.Sound("backsound.wav")
eat_p_sound = pygame.mixer.Sound("eat_p.wav")
eat_m_sound = pygame.mixer.Sound("eat_m.wav")
hit_sound = pygame.mixer.Sound("hit.wav")
dead_sound = pygame.mixer.Sound("dead.wav")
heal_sound = pygame.mixer.Sound("Item_heal.wav")
slow_sound = pygame.mixer.Sound("Item_slow.wav")
fast_sound = pygame.mixer.Sound("Item_fast.wav")
stop_sound = pygame.mixer.Sound("Item_stop.wav")


pygame.mouse.set_visible(0)

plus_list = pygame.sprite.Group()
all_sprites_list =pygame.sprite.Group()

player = Player()
all_sprites_list.add(player)


#타이머
total_time = 10
start_ticks = pygame.time.get_ticks()
font = pygame.font.Font("malgun.ttf", 25)
point = 0
level = 1


clock = pygame.time.Clock()
game_over = False
exit_program = False


while exit_program != True:
    clock.tick(30)
    screen.fill(white)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            exit_program=True
    if not game_over:
        player.update()
        game_over = plus.update(-5)
    if game_over:
        text=font.render("Game Over", 1, black)
        textpos = text.get_rect(centerx = screen.get_widht()/2)
        screen.blit(text, textpos)
        
    
    #배경 넣기
    screen.blit(background_image, background_position)
    
    #마우스 커서
    player_position = pygame.mouse.get_pos()
    x = player_position[0]
    y = player_position[1]
        
    screen.blit(player_image, [x,y])
    
    
    
    rand_many = randrange(5)
    for i in range(int(rand_many)):
        plus = Plus(0,0)
        plus.rect.x = randrange(0,screen_width)
        plus.rect.y = randrange(-130,-30)
        plus_list.add(plus)
        if plus.rect.y != screen_height:
            plus.rect.y += 30
        
        all_sprites_list.add(plus)
        
    all_sprites_list.draw(screen)
    
    
    text = font.render("Point: " + str(point), True, black)
    screen.blit(text, [10,40])
    
    text = font.render("Level: " + str(level), True, black)
    screen.blit(text, [10,70])
    
    plus.update(level)
    
    
    
    #경과시간 계산
    elapsed_time = (pygame.time.get_ticks() - start_ticks) / 1000
    timer = font.render("Timer: " + str(int(elapsed_time)), True, black)

    #경과 시간 표시
    screen.blit(screen, (0, 0))
    screen.blit(timer, (10, 10))
    

    pygame.display.flip()

pygame.quit()
```
