## Pylnvader 프로그램2

```python
import pygame
import random

white = (255, 255, 255)
black = (0, 0, 0)
red = (255, 0, 0)
green = (0, 255, 0)
blue = (0, 0, 255)

class Sprite:
    def __init__(self, xpos, ypos, filename):
        self.x = xpos
        self.y = ypos
        self.bitmap = pygame.image.load(filename)
        self.bitmap.set_colorkey(white)

    def set_position(self, xpos, ypos):
        self.x = xpos
        self.y = ypos

    def render(self):
        screen.blit(self.bitmap, (self.x, self.y))

def Collision(x1, y1, x2, y2):
    if (x1 > x2-30) and(x1 < x2+30) and (y1 > y2-30) and (y1 < y2+30):
        return True
    else:
        return False

pygame.init()

#가독성을 향상시키기 위해
screen_width = 640
screen_height = 480
screen = pygame.display.set_mode([screen_width,screen_height])
pygame.display.set_caption('Pylnvaders2')
pygame.key.set_repeat(1, 1)
font = pygame.font.Font('malgun.ttf', 36)

#크기조절한 폰트 추가
font2 = pygame.font.Font('malgun.ttf', 26)
backdrop = pygame.image.load('back.png')

#효과음 및 포인트, 레벨, 타이머 초기값선언
load_sound = pygame.mixer.Sound("load.wav")
shoot_sound = pygame.mixer.Sound("shoot.wav")
point = 0
level = 1
total_time = 10
start_ticks = pygame.time.get_ticks()
god_mode = 0

#우주선 이미지, 미사일 이미지 필요
hero = Sprite(20, 400, 'hero_player.bmp')
ourmissile =Sprite(0, 480, 'heromissile_player.bmp')


#적 이미지, 미사일 이미지 필요
enemies = []
x = 0
for count in range(10):
    enemies.append(Sprite(50 * x + 50, 50, 'baddie_enemy.bmp'))
    x += 1

#미사일을 여러개 만들기 위해 리스트화
enemymissile = []
for count in range(1,2):
    enemymissile_x = enemies[random.randint(0, len(enemies) -1)].x
    enemymissile_y = enemies[0].y
    enemymissile.append(Sprite(enemymissile_x, enemymissile_y, 'baddiemissile_enemy.bmp'))




enemyspeed = 3
clock = pygame.time.Clock()
quit = False
game_over = False

while quit != True:
    screen.blit(backdrop, (0, 0))
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            quit = True
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_RIGHT and hero.x < 590:
                hero.x += 5
            if event.key == pygame.K_LEFT and hero.x > 10:
                hero.x -= 5
            if event.key == pygame.K_SPACE:
                ourmissile.x = hero.x
                ourmissile.y = hero.y
                shoot_sound.play()
                
        #아래버튼을 관리자 모드로 들어가며 
        #DOWN버튼을 누르면 자동으로 적이 사라진다.     
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_DOWN:
                god_mode = 1
                for count in range(0, len(enemies)):
                    point += 1
                    del enemies[count]
                    break
            
    #만약 DOWN버튼을 한번이라도 실행했을 시 "관리자 모드"라는 메시지가 나타남        
    if god_mode == 1:
        text = font2.render("관리자 모드", True, red)
        screen.blit(text, [500,40])
    
    if not game_over:
        
        #DOWN을 눌러 적을 사라지게 할 시 이부분이 밑에 있다면 오류가 나타난다
        #레벨에 따라서 적이 늘어나도록 조정
        #10,20,30,40을 통과하면 적의 수(열)이 증가하도록 하였음
        if len(enemies) == 0:
            x = 0
            y = 0
            z = 0
            level += 1
            
            if level >= 40:
                y = 5      
            elif level >= 30:
                y = 4
            elif level >= 20:
                y = 3  
            elif level >= 10:
                y = 2
            else:
                y = 1
                
            for row in range(y):
                    z += 1
                    x = 0
                    if z%2 == 0:
                        for count in range(10):
                            enemies.append(Sprite(50 * x + 25, z * 50, 'baddie_enemy.bmp'))
                            x += 1
                    
                    elif z%2 != 0:
                        for count in range(10):
                            enemies.append(Sprite(50 * x + 50, z * 50, 'baddie_enemy.bmp'))
                            x += 1
                
                
        for count in range(len(enemies)):
            enemies[count].x += enemyspeed
        
        if enemies[len(enemies)-1].x > 590:
            enemyspeed = -3
            for count in range(len(enemies)):
                enemies[count].y += 5
        
        if enemies[0].x < 10:
            enemyspeed = 3
            for count in range(len(enemies)):
                enemies[count].y += 5
        
        
        #만약 적들이 끝까지 도달하면 게임오버        
        for count in range(len(enemies)):
            if enemies[count].y >= 480:
                game_over = True

        if ourmissile.y <479 and ourmissile.y > 0:
            ourmissile.render()
            ourmissile.y -= 5
        
        #레벨 증가에 따른 미사일 갯수 증가
        for count in range(len(enemymissile)):
            if enemymissile[count].y >= 480 and len(enemies) > 0:
                del enemymissile[count]
                
                for count in range(level):
                    enemymissile_x = enemies[random.randint(0, len(enemies) -1)].x
                    enemymissile_y = enemies[0].y
                    enemymissile.append(Sprite(enemymissile_x, enemymissile_y, 'baddiemissile_enemy.bmp'))
                break
                
                
        
                
        for count in range(len(enemymissile)):    
            if Collision(hero.x, hero.y, enemymissile[count].x, enemymissile[count].y):
                game_over = True
            
        #포인트 1이 추가 되며 효과음 재생 및 count번의 리스트 삭제
        for count in range(0, len(enemies)):
            if Collision(ourmissile.x, ourmissile.y, enemies[count].x, enemies[count].y):
                point += 1
                load_sound.play()
                del enemies[count]
                break        
        
        for count in range(len(enemymissile)):
            enemymissile[count].render()
        
        #레벨에 따른 미사일의 속도 조정
        for count in range(len(enemymissile)):
            if level >= 45:
                enemymissile[count].y += 15
            elif level >= 35:
                enemymissile[count].y += 11
            elif level >= 25:
                enemymissile[count].y += 9
            elif level >= 15:
                enemymissile[count].y += 7
            elif level >= 7:
                enemymissile[count].y += 5
            elif level >= 3:
                enemymissile[count].y += 3
            else:
                enemymissile[count].y += 1

            
    if game_over:
        text = font.render("Game Over", 1, red)
        screen.blit(text, [200, 200])

    hero.render()

    for count in range(len(enemies)):
        enemies[count].render()
    
    
        
    #레벨 표시
    text = font2.render("Level : " + str(level), True, green)
    screen.blit(text, [515,10])
               
    #포인트 계산
    text = font2.render("Point : " + str(point), True, green)
    screen.blit(text, [10,40])
 
    #경과시간 계산
    elapsed_time = (pygame.time.get_ticks() - start_ticks) / 1000
    timer = font2.render("Timer : " + str(int(elapsed_time)), True, green)

    #경과 시간 표시
    screen.blit(screen, (0, 0))
    screen.blit(timer, (10, 10))    
        
    clock.tick(70)
    pygame.display.update()

pygame.quit()


```
