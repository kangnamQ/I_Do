## 계단 오르내리기 예제 - 파이썬

```python
import random

position = 0
walk = []
steps = 100

for i in range(steps):
    if random.randint(0, 1):  #<< 0~1인 난수 발생
        step = 1
    else:
        step = -1
    position += step
    walk.append(position)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = range(steps) #<< 0~99 까지 100번 반복 (step이 100이라)
plt.title("Random walk with +1/-1 steps")
plt.plot(x, walk)
plt.show()
```
