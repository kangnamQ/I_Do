## 마우스로 공 피하기

```python
import pygame

black = [0,0,0]
white = [255,255,255]
blue = [0,0,255]
green = [0,255,0]
red = [255,0,0]

#공
class Ball():
    x = 0
    y = 0
    change_x = 0
    change_y = 0
    size = 10
    color = white
    
    def draw(self,screen):
        pygame.draw.circle(screen, self.color,[self.x,self.y],self.size)
        
    def move(self):
        self.x += self.change_x
        self.y += self.change_y
        
    def bounce(self):
        if self.y > 300 or self.y < 10:
            self.change_y = self.change_y * -1
        
        if self.x > 400 or self.x < 10:
            self.change_x = self.change_x * -1

#마우스           
def draw_stick_figure_m(screen,x1,y1):
    #머리
    pygame.draw.ellipse(screen,black,[1+x1,y1,10,10],0)
    #몸    
    pygame.draw.line(screen,black,[5+x1,17+y1],[5+x1,7+y1],2)
    #팔
    pygame.draw.line(screen,black,[5+x1,7+y1],[9+x1,17+y1],2)
    pygame.draw.line(screen,black,[5+x1,7+y1],[1+x1,17+y1],2)
    #다리
    pygame.draw.line(screen,black,[5+x1,17+y1],[10+x1,27+y1],2)
    pygame.draw.line(screen,black,[5+x1,17+y1],[x1,27+y1],2)

#키보드           
#def draw_stick_figure_k(screen,x2,y2):
#    #머리
#    pygame.draw.ellipse(screen,black,[1+x2,y2,10,10],0)
#    #몸    
#    pygame.draw.line(screen,black,[5+x2,17+y2],[5+x2,7+y2],2)
#    #팔
#    pygame.draw.line(screen,black,[5+x2,7+y2],[9+x2,17+y2],2)
#    pygame.draw.line(screen,black,[5+x2,7+y2],[1+x2,17+y2],2)
#    #다리
#    pygame.draw.line(screen,black,[5+x2,17+y2],[10+x2,27+y2],2)
#    pygame.draw.line(screen,black,[5+x2,17+y2],[x2,27+y2],2)
    
pygame.init()
        
        
size=[400,300]
screen=pygame.display.set_mode(size)
pygame.display.set_caption("Avoid the ball")



done = False
clock = pygame.time.Clock()

#마우스 포인터 안보이게
pygame.mouse.set_visible(0)

#공
redBall = Ball()
blueBall = Ball()
greenBall = Ball()
blackBall = Ball()

redBall.x = 10
redBall.y = 10

blueBall.x = 400
blueBall.y = 300

greenBall.x = 200
greenBall.y = 100

blackBall.x = 300
blackBall.y = 200


redBall.change_x = 5
redBall.change_y = 5

blueBall.change_x = -7
blueBall.change_y = -7

greenBall.change_x = 9
greenBall.change_y = 9

blackBall.change_x = -11
blackBall.change_y = -11

redBall.color = red
blueBall.color = blue
greenBall.color = green
blackBall.color = black


#키보드
#x2_speed = 0
#y2_speed = 0

#x2_coord = 10
#y2_coord = 10

#타이머
total_time = 10
start_ticks = pygame.time.get_ticks()
font = pygame.font.Font("malgun.ttf", 25)

T = 10

while done==False:
    for event in pygame.event.get():
        dt = clock.tick(60)
        if event.type == pygame.QUIT:
            done=True
            
#        if event.type == pygame.KEYDOWN:
#            if event.key == pygame.K_LEFT:
#                x_speed = -3
#            if event.key == pygame.K_RIGHT:
#                x_speed = +3
#            if event.key == pygame.K_UP:
#                y_speed = -3
#            if event.key == pygame.K_DOWN:
#                y_speed = +3
                         
#        if event.type == pygame.KEYUP:
#            if event.key == pygame.K_LEFT:
#                x_speed = 0
#            if event.key == pygame.K_RIGHT:
#                x_speed = 0
#            if event.key == pygame.K_UP:
#                y_speed = 0
#            if event.key == pygame.K_DOWN:
#                y_speed = 0
  
    screen.fill(white)
    
    
    redBall.draw(screen)
    blueBall.draw(screen)
    
    redBall.move()
    blueBall.move()
    
    redBall.bounce()
    blueBall.bounce()
     
    #마우스
    pos = pygame.mouse.get_pos()
    x1=pos[0]
    y1=pos[1]
    draw_stick_figure_m(screen,x1,y1)  
    
    #키보드
#    x2_coord = x2_coord+x2_speed
#    y2_coord = y2_coord+y2_speed    

#    draw_stick_figure_k(screen,x2_coord,y2_coord)    


    #경과시간 계산
    
    elapsed_time = (pygame.time.get_ticks() - start_ticks) / 1000
    
    timer = font.render("Level : " + str(int(T/10)) + " Timer: " + str(int(total_time - elapsed_time)), True, black)

    #경과 시간 표시
    screen.blit(screen, (0, 0))
    screen.blit(timer, (10, 10))
    
    if total_time - elapsed_time <= 0:
        print("조금 더 빠르게!")
        total_time = total_time + 10
        T += 10 
        
    if T > 5:
        greenBall.draw(screen)  
        greenBall.move()
        greenBall.bounce()
        
        
    if T > 10:
        blackBall.draw(screen)  
        blackBall.move()
        blackBall.bounce()
        
    clock.tick(T)
    
    pygame.display.update()
    pygame.display.flip()
    
pygame.quit()
```
